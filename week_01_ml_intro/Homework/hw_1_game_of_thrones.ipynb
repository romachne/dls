```python
#   ,     
#      .
#  ,     

d = {'A': ['a1', 'a2', 'a3'],
     'B': ['b1', 'b2', 'b3', 'b4']}

d_inverted = {}
for k in d.keys():
  for v in d[k]:
      d_inverted.update({v:k})

d_inverted
```

```python
#        cultures_grouped
# cultures_grouped_inverted = ...
# ...
cultures_grouped_inverted = {}
for k in cultures_grouped.keys():
    for v in cultures_grouped[k]:
        cultures_grouped_inverted.update({v:k})

cultures_grouped_inverted
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(
    handle_unknown='ignore',
    sparse_output=False,
    dtype=int
)

categorical_features = [
    'culture_grouped',
    'house',
]

encoded = encoder.fit_transform(data[categorical_features])
one_hot_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_features), index=data.index)

data_encoded = pd.concat([data, one_hot_df], axis=1)
data_encoded = data_encoded.drop(categorical_features, axis=1)
data_encoded.drop(columns=['title', 'name', 'culture', 'mother', 'father', 'heir', 'spouse'], inplace=True)
data_encoded = data_encoded.fillna(0)
data_encoded
```

** 1.9.**   `X`,      ,       ,  `y`,       .

```python
X = data_encoded.drop(columns=['isAlive'])
y = data_encoded['isAlive']
```

** 1.10.**       (train  val)    `train_test_split` (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

```python
from sklearn.model_selection import train_test_split
```

```python
#     train_test_split   random_state,
#       train  validation .
#          
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2, random_state=44)
```
